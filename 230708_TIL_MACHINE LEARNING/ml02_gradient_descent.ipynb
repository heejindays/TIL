```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# x : weight
# y : mse

x = np.arange(-10, 11, 1)
y = x ** 2
```

```python
# 평균 제곱 오차(MSE) : 예측값과 실제값의 차이를 제곱한 후 평균

# 오차의 제곱을 가지고 그래프를 그림 (이차함수 그래프)
# 오차의 제곱을 미분하면 기울기가 나오고 1차 함수
# 특정 점에서 기울기를 구함

# 특정 점에서의 기울기 (미분값) : x에서의 y의 변화량이 기울기

# 오차제곱평균이 작아질수록 정답에 가까움
# 기울기가 줄어들고 있다는 것은 MSE가 작아진다 (정답에 가까워지고 있다)
# 기울기가 0이라면 정답이다

plt.plot(x, y)
plt.show()
```

```python
# Gradient Descent 경사 하강법
# 기울기가 점점 줄어들게 만들고 싶다 > MSE를 줄이겠다 > 정답에 가까워지고 싶다
# w랑 b를 정답에 가까워지게 하겠다
```

```python
# 기울기
x_new = 10

# 기울기에 맞는 x값
cost = []

# 기울기에 맞는 mse값 (y값)
hy = []

# 기울기를 움직이는 비율
learning_rate = 0.1
```

```python
for i in range(100):
    
    # 기울기를 옛날 기울기로 바꿈
    old_value = x_new

    # 기울기를 0.1 만큼 곱해서 빼버려서 새롭게 기울기를 만듦 (기울기가 조금 줄어듦)
    cost.append(old_value - learning_rate * old_value)
    x_new = old_value - learning_rate * old_value

    # 기울기에 제곱을 해서 append 해줌
    # (간단한 의미로 적분했다고 생각하면 됨)
    # learning_rate 만큼 이동함 > 진행될수록 0에 가까워짐
    # 학습을 한번 할 때마다 기울기를 어떻게 움직일 것인가
    # learning_rate = 0.1는 경험적으로 정해야 함
    hy.append(x_new ** 2)
```

```python
plt.plot(x, y)
plt.scatter(cost, hy)
plt.show()
```
