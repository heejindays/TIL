```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-5, 5, 0.1)
```

```python
# dl01에서 만든 활성화 함수의 이름이 binary_step

def binary_step(x):
    return np.array(x > 0)
```

```python
plt.plot(x, binary_step(x))
```

```python
def gaussian(x, mu=0, sigma=1, a=1):
    return a * np.exp(-(x - mu) ** 2 / (2 * sigma ** 2))
```

```python
plt.plot(x, gaussian(x))
plt.plot(x, gaussian(x, 2, 3, 0.5))
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
plt.plot(x, sigmoid(x))

# 시그모이드 함수는 0부터 1까지로 나옴
```

```python
def relu(x):
    return np.maximum(0, x)
```

```python
plt.plot(x, relu(x))

# 0보다 작으면 0이고 0보다 크면 x
```

```python
def softmax(x):
    return np.exp(x) / np.sum(np.exp(x))
```

```python
x = np.array([0.3, 2.9, 4.0])
y = softmax(x)
```

```python
print(y)
print(y.sum())
```
