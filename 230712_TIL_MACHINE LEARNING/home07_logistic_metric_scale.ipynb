```python
import pandas as pd
```

```python
df = pd.read_csv("../datasets/weight_height.csv", encoding="euc-kr")
df
```

```python
df = df[["학교명", "학년", "성별", "키", "몸무게"]]
df
```

```python
df["grade"] = list(map(lambda x:0 if x[-4] == "초등학교"
                       else 6 if x[-3] == "중학교"
                       else 9, df["학교명"])) + df["학년"]
df
```

```python
# "학교명"과 "학년" 열은 'grade'를 만든 뒤에는 쓸모 없으니 삭제
df.drop(["학교명", "학년"], axis="columns", inplace=True)
df
```

```python
# 컬럼 이름 바꾸기
df.columns = ["gender", "height", "weight", "grade"]
df
```

```python
# 데이터를 스케일링 후 학습!

# 몸무게와 키와 성별을 학습시켜서
# 몸무게와 키로 성별을 예측하자 (분류)
# 남자 : 0, 여자 : 1 형태로 변환

# 평가 지표 사용해보기!
# confusion matrix, presision, recall, roc_curve, roc_auc_score
```

```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, precision_recall_curve, precision_score, recall_score, f1_score, accuracy_score, classification_report, precision_recall_curve, roc_curve, roc_auc_score
```

```python
# 성별을 0 또는 1로 바꾸자
df["gender"] = df["gender"].map(lambda x: 0 if x == "남" else 1)
df
```

```python
# 결측치 제거

# df.isna().sum()
df.dropna(inplace=True)
df.isna().sum()
```

```python
# 1. 학습할 데이터 준비

x = df[["weight", "height"]]
y = df[["gender"]]
```

```python
x_scale = StandardScaler().fit_transform(x)
```

```python
# 데이터 나누기

x_train, x_test, y_train, y_test = train_test_split(x_scale, y, test_size=0.3, random_state=1)
```

```python
# 모델 선택

model = LogisticRegression()
```

```python
# 학습
model.fit(x_train, y_train)
```

```python
# 평가
model.score(x_test, y_test)
```

```python
# 6. 예측
predict = model.predict(x_test)
predict
```

```python
cm = confusion_matrix(y_test, predict)
cm
```

```python
ConfusionMatrixDisplay(cm).plot()
```

여학생이 1, 남학생이 0

- TP : 여학생이라고 예측 > 여학생이 맞음 (580)

- TN : 남학생이라고 예측 > 남학생이 맞음 (815)

- FP : 여학생이라고 예측 > 남학생이었음 (553)

- FN : 남학생이라고 예측 > 여학생이었음 (957)

- TPR : True Positive Rate

- FPR : False Positive Rate

```python
TN, FP, FN, TP = cm.ravel()
girl_precision = TP / (TP + FP)
girl_recall = TP / (TP + FN)
print(girl_precision)
print(girl_recall)
```

```python
precision = precision_score(y_test, predict)
recall = recall_score(y_test, predict)
print(precision)
print(recall)
```

```python
my_f1_score = 2 * ((girl_precision * girl_recall) / (girl_precision + girl_recall))
print(my_f1_score)
```

```python
f1_score(y_test, predict)
```

```python
proba = model.predict_proba(x_test)
proba
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve
```

```python
precision, recall, thresholds = precision_recall_curve(y_test, proba[:,1], pos_label=1)

plt.xlabel("threshold")
plt.ylabel("score")

plt.plot(thresholds, precision[:len(thresholds)], "red", linestyle="--", label="presicion")

plt.plot(thresholds, recall[:len(thresholds)], "blue", label="recall")

plt.legend()
plt.show()
```

threshold를 기준으로

높아지면 : presicion 높아짐 (예측한 여학생 중에서 실제 여학생일 비율이 올라감) > FP 낮아짐 > FPR이 낮아짐 (FPR = FP / (FP + TN) 이니까)

낮아지면 : recall 높아짐 (전체 여학생에서 실제 여학생 비율) > FPR이 높아짐


딱 맞추는 것이 중요한지 vs 아니면 조금 틀려도 되니까 최대한 포함되게 하는게 중요한지


1) 여학생을 뽑아서 목욕탕에 보낼 것인데 이럴 때 threshold를 어디로 보내야 할까 (꼭 여자여야 함)

> 오른쪽 (FP 낮아지도록 : 사실은 남학생인 비율을 줄이자)


2) 여학생들에게 립스틱을 지원해줄 것 threshold 어디로? (남자도 받아도 됨)

> 왼쪽 (전체 여학생들이 최대한 받으라고 : 남학생들이 받아도 괜찮음 / 여학생들이 최대한 많이 받으면 됨)


- threshold INC -> precision inc(FP dec), recall dec(FN inc, FPR dec)

- threshold DEC -> precision dec(FP inc), recall inc (FN dec) -> high FPR, TPR

```python
fprs, tprs, thresholds = roc_curve(y_test, proba[:, 1], pos_label=1)

plt.plot(fprs, tprs)
plt.plot([0, 1], [0, 1], "--")

plt.xlabel("fpr")
plt.ylabel("tpr")
plt.show()
```

```python
# multicalss 속성 (default : raise)
# ovr : one vs rest > 클래스가 불균형할 때
# ovo : one vs one  > 가능한 모든 클래스 조합

roc_auc_score(y_test, proba[:, 1])
```
