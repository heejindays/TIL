```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler, normalize
```

데이터의 단위가 다르면 숫자가 큰 것들에 끌려다닐 수 밖에 없음

어떻게 하면 단위를 맞출 수 있을까?

[스케일링]

1) MinMaxScaler : 0 ~ 1

2) StandardScaler : 평균은 0, 분산은 1 (정규분포 형태)

3) Normalizer : 0 ~ 1 (거리니까)

각각 사용하는 종류가 다름

```python
np.random.seed(100)
data_array = list()

for i in range(1, 100):
    r = np.random.randint(0, i*10, 10)
    data_array.extend(r)

data_array.extend(np.zeros(100))
data_array
```

```python
data = pd.DataFrame({"scaling_data": data_array})
print(data)
```

```python
# 0부터 977까지의 값이 만들어져 있음

print(f"max : {data.max()}")
print(f"min : {data.min()}")
```

```python
# 선형회귀 때 주로 사용
# MinMaxScaler() : 0과 1을 포함함
scaler = MinMaxScaler()
```

```python
data_n = scaler.fit_transform(data)
data_n
```

```python
print(data_n.min())
print(data_n.max())
```

```python
# 분류할 때 주로 사용
# 정규분포를 따른다는 가정 하에 정규분포 형식으로 바꿈
scaler = StandardScaler()
data_n = scaler.fit_transform(data)
data_n
```

```python
# 유클리드 거리 치환 -> 딥러닝
data_n = normalize([data["scaling_data"]], norm="l2")
data_n
```
