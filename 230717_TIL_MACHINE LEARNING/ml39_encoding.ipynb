```python
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
tips.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
tips["time_encoded"] = LabelEncoder().fit_transform(tips["time"])
print(tips["time"].value_counts())
print(tips["time_encoded"].value_counts())

# 디너는 0, 런치는 1로 바꿈
# 숫자가 아닌 것들을 숫자값으로 바꾸는 것이 라벨 인코더
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
tips["day"].value_counts()

"""
Sat     87
Sun     76
Thur    62
Fri     19

이렇게 4개가 있어서 그 중에서 하나만 1을 주고 나머지는 0
"""
```

```python
# OneHotEncoder() 중요
# 데이터를 준비하는 단계에서 진행
tips_day = OneHotEncoder().fit_transform(tips[["day"]])
tips_day.toarray()
```

```python
tips_day_df = pd.DataFrame(tips_day.toarray())
tips_day_df["day"] = tips["day"]
tips_day_df
```
