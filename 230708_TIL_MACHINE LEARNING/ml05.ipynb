```python
import pandas as pd
import numpy as np
```

```python
# 보스턴 집 값 예제 데이터 가져오기
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]
```

```python
boston = pd.DataFrame(data, columns=["CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE", "DIS", "RAD", "TAX", "PTRATIDO", "B", "LSTAT"])

# "B"는 흑인 거주 비율이어서 예제 데이터에서 빠짐
```

```python
boston
```

```python
boston["PRICE"] = target
boston
```

- CRIM : 지역별 범죄 발생 비율
- ZN : 25,000 평방피트를 초과하는 거주 지역 비율
- NDUS : 비상업 지역의 토지 비율
- CHAS : 찰스 강에 대한 더미 변수 (강의 경계에 인접:1 아니면:0)
- NOX : 일산화질소 농도
- RM : 평균 방 갯수
- AGE : 1940년 이전에 건축된 소유 주택 비율
- DIS : 5개 주요 고용센터까지의 접근성
- RAD : 고속도로 접근성
- TAX : 10,000 달러 당 재산 세율
- PTRATIO : 지역별 교사와 학생 수 비율
- B : 지역별 흑인 거주 비율
- LSTAT : 하위 계층 비율
- PRICE : 주택 가격(중앙값)

```python
# RM으로 PRICE 에측하기

# 1. 데이터 준비
# x : RM, y : PRICE

x = boston[["RM"]]
y = boston[["PRICE"]]
```

```python
print(x)
print(y)
```

```python
# 2. 데이터 분할 (trainset, testset)

from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
# 3. 모델 선택
from sklearn.linear_model import LinearRegression
```

```python
# 연속적인 값이니까 선형
model = LinearRegression()
```

```python
# 4. 모델 학습 (trainset)
model.fit(x_train, y_train)
```

```python
# 5. 예측 (testset)
predict = model.predict(x_test)
predict
```

```python
# 어느정도 오차가 있는지 궁금해
result = y_test
result["predict"] = predict
result
```

```python
from sklearn.metrics import mean_squared_error
```

```python
# 결과를 보고 어느정도 오차가 있는지 알고 싶었음
# 결과를 보니까 36 정도 나옴
from sklearn.metrics import mean_squared_error
mean_squared_error(result["PRICE"], result["predict"])
```

```python
# 예측한 값과 실제 값의 정확도를 판별해보고 싶음
# price는 실제 있는 값이고 predict는 X_test를 가지고 만든 것
# score는 정확도
# 지금 60% 정도니까 이거 하나만 가지고는 확인이 어렵겠다고 판단할 수 있음
model.score(x_test, y_test["PRICE"])
```

```python
# 5. 예측 (testset)
predict = model.predict(x_test)
predict
```

```python
model.score(x_test, y_test)
```

```python
predict = model.predict(x_test)
```

```python
model.score(x_test, y_test)
# 그냥 상관관계가 높은 걸 하는게 아니고 X를 다 때려 넣어봤더니 더 높은 확률로 맞힘
```
