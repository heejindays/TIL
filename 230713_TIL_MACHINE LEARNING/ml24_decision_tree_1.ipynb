Decision Tree (의사 결정 나무)

- 스무 고개 놀이처럼 동작

- 조건에 대한 참과 거짓을 물어보면서 맞혀 나가는 과정

- 기준을 생성하고 True / False 과정을 거쳐 뻗어 나감

- 학습을 통해 트리 형태의 기준을 생성

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
cancer
```

```python
cancer.data
```

```python
cancer.target
```

```python
cancer.feature_names
```

```python
# 0 : malignant 악성
# 1 : benign 양성
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(cancer.data, columns=cancer.feature_names)
df
```

```python
# 컬럼명 사이에 있는 공백을 없애고 싶다

df.columns = df.columns.map(lambda x: "_".join(x.split(" ")))
df
```

```python
df.info()
```

```python
df["target"] = cancer.target
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(25, 15))
for idx, column in enumerate(df.iloc[:, :-1]):
    plt.subplot(6, 5, idx+1)
    sns.histplot(data=df, x=column, hue="target")

plt.tight_layout()
plt.show()
```

```python
mal = df[df["target"] == 0]
ben = df[df["target"] == 1]

print(f"악성 :{mal.shape}")
print(f"양성 :{ben.shape}")
```

```python
# 악성인 것들 (타겟이 0)
mal
```

```python
# 양성인 것들 (타겟이 1)
ben
```

```python
# 악성 히트맵

plt.figure(figsize=(15, 15))
sns.heatmap(mal.corr(), annot=True)
```

```python
# 양성 히트맵
plt.figure(figsize=(15, 15))
sns.heatmap(ben.corr(), annot=True)
```

```python
x = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
# 스코어가 높다고 다 좋은 게 아님 > 과적합 되어있을 수도 있음
```

```python
# 나무의 깊이를 4까지만으로 설정해보자
# max_depth= 중요한 파라미터
model = DecisionTreeClassifier(max_depth=4)
model.fit(x_train, y_train)
model.score(x_test, y_test)
```

```python
from sklearn.tree import export_graphviz
```

```python
"""
impurity : 균일도 / 불순도
- 규칙조건 생성에서 중요한 것
- 정보 균일도가 높은 데이터 셋을 먼저 선택할 수 있도록 규칙 조건을 만드는 것이 중요하다
- impurity=False: 불순도 지표(impurity)를 표시하지 않도록 설정

filled : True로 설정하면 분류
filled=True: 트리 노드를 불순도에 따라 색으로 채움
"""

export_graphviz(model, out_file="tree dot", class_names=["악성", "양성"], feature_names=cancer.feature_names, impurity=False, filled=True)

export_graphviz(model)
```

```python
from sklearn.tree import plot_tree
```

```python
# 사이킷런 안에서도 만들 수 있음

plt.figure(figsize=(25, 10))
plot_tree(model, class_names=["mal", "ben"],
          feature_names=list(cancer.feature_names),
          impurity=False,
          filled=True,
          fontsize=10)

plt.show()
```

```python
# 특성 중요도 : 트리를 결정하는데 각 특성이 얼마나 중요한지를 확인
# 중요도가 낮다고 해서 필요 없는 것이 아님
# 트리가 선택하지 않았을 뿐
# (다른 특성이 동일한 정보를 지니고 있어서일 가능성도 있음)

# 낮은 특성 중요도는 해당 특성이 트리 모델에서 분할에 활용되지 않았거나 
# 다른 특성과 상호작용이 적었음을 의미
model.feature_importances_
```

```python
sns.barplot(x=model.feature_importances_, y=df.columns[:-1])
# worst_perimeter가 다른 것들에 비해서 많이 쓰였다
```
