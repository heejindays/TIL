```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
```

```python
x, y = make_blobs(random_state=1)
```

```python
predict = AgglomerativeClustering(n_clusters=3).fit_predict(x)

# 병합 군집
# 하나 중심을 잡고 근처에 있는 것들을 하나씩 연결해서 묶어 나감
# 그다지 중요하지 않은 분류 분석
```

```python
plt.scatter(x[:, 0], x[:, 1], c=predict)
```

```python
from scipy.cluster.hierarchy import dendrogram, ward
```

```python
x, y = make_blobs(random_state=0, n_samples=10)

linkage_array = ward(x)
linkage_array

# ward : 거리 정보가 담긴 배열
```

```python
dendrogram(linkage_array)

ax = plt.gca()
bounds = ax.get_xbound()

ax.plot(bounds, [7.5, 7.5], "--", c="k")
ax.plot(bounds, [5.5, 5.5], "--", c="k")

ax.text(bounds[1], 7.5, "n_cluster:2", va="center")
ax.text(bounds[1], 5.5, "n_cluster:3", va="center")
```
