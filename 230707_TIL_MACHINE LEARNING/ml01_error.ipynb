```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 입력된 값 (문제(x값) / 정답(y값))
# 이 데이터를 가지고 컴퓨터한테 x와 y를 직접 알아내라고 하고 싶다
# 궁극적인 목표! x 값을 주면 y 값을 도출하게 만들고 싶다 (x와 y의 관계를)
x = np.array([1, 2, 3, 4, 5])
y = np.array([3, 4, 10, 14, 17])
```

```python
plt.plot(x, y)
plt.show()
```

```python
# 우리가 컴퓨터라고 생각하고 식을 만들어 본 것
# w : 3 / b : 1
fx01 = 3 * x + 1
fx01
```

```python
fx01_error = y - fx01
# 실제 값에서 예측한 값을 빼서 확인 중

print(fx01_error)
print(np.sum(fx01_error))
```

```python
plt.plot(x, fx01)
```

```python
# w * x + b > 0 * x + 10
# w:0 / b:10

fx02 = 0 * x + 10
fx02

# a랑 b에 10을 넣었다고 치자
```

```python
fx02_error = y - fx02
print(fx02_error)
print(np.sum(fx02_error))
```

```python
plt.plot(x, fx02)
for i in range(len(x)):
    hx = np.array([x[i], x[i]])
    hy = np.array([fx02[i], y[i]])
    # (x[i], fx02[i]) 으로 찍힘 (선 위)
    # (fx02[i], y[i]) 으로 찍힘
    plt.plot(hx, hy, marker="o")
plt.show()
```

정답과 예측의 차이를 단순하게 더하면 판단을 제대로 내릴 수 없음

오차들을 모두 제곱해서 더해서 평균을 내자

-> Mean Squared Error (평균 제곱 오차)

```python
def mse(a, b, x, y):
    return (((y - (a*x+b))**2)).mean()
```

```python
# 오차의 제곱의 평균을 내보니까 이게 더 작음
# 첫번째가 더 실제와 가깝다는 걸 알 수 있음
# 3x + 1이 0x + 10보다 오차가 작구나
# 오차가 점점 작아지게 만드는 것 -> 학습

print(mse(3, 1, x, y))
print(mse(0, 10, x, y))
```
