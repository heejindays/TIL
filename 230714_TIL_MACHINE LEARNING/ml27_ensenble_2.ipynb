```python
from sklearn.datasets import make_moons, load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# make_moons : 연습용 데이터 생성
x, y = make_moons(n_samples=1000, noise=0.25, random_state=1)
```

```python
plt.plot(x[y==1, 0], x[y==1, 1], marker="o", color="red", linestyle="")
plt.plot(x[y==0, 0], x[y==0, 1], marker="o", color="blue", linestyle="")

plt.show()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=1)
```

```python
model = RandomForestClassifier(n_estimators=5)
```

```python
model.fit(x_train, y_train)
```

```python
predict = model.predict(x_test)
accuracy_score(y_test, predict)
```

```python
# 앙상블 모델(Ensemble model)에서 개별 모델의 리스트를 반환하는 속성
model.estimators_
```

```python
cancer = load_breast_cancer()
```

```python
x_train, x_test, y_train, y_test = train_test_split(cancer.data, cancer.target, test_size=0.3, random_state=1)
```

```python
model = RandomForestClassifier()
```

```python
model.fit(x_train, y_train)
```

```python
predict = model.predict(x_test)
accuracy_score(y_test, predict)
```

```python
sns.barplot(x=model.feature_importances_, y=cancer.feature_names)
```
